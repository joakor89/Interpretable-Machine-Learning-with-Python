## Understanding a Simple Weight Prediction Model

### Loading & Preparing the Data

#### Loading Webpage & Extracting Content into DataFrame

```python
# url = 'http://wiki.stat.ucla.edu/socr/index.php/SOCR_Data_Dinov_020108_HeightsWeights'

# height_weight_df = pd.read_html(url)[1][['Height(Inches)','Weight(Pounds)']]
```

```python
url = 'https://people.sc.fsu.edu/~jburkardt/data/csv/hw_200.csv'

df = pd.read_csv(url)

print(df.columns)
```

```python
# Cleaning Column Names 
df.columns = df.columns.str.strip().str.replace('"', '')

# Column Slection
height_weight_df = df[['Height(Inches)', 'Weight(Pounds)']]
```

```python
print(height_weight_df.head())
```

#### Counting Records

```python
num_records = len(height_weight_df)
print(num_records)
```

#### Placing `x & y` Variables

```python
x = height_weight_df['Height(Inches)'].values.reshape(num_records, 1)

y = height_weight_df['Weight(Pounds)'].values.reshape(num_records, 1)
```

### A Linear Regression Model

#### Fitting Model

```python
model = linear_model.LinearRegression().fit(x,y)
```

#### Generating Equation

```python
print("ŷ = " + str(model.intercept_[0]) + " + " + str(model.coef_.T[0][0]) + " x₁")
```

#### Computing Mean Absolute Error

```python
y_pred = model.predict(x)

mae = mean_absolute_error(y, y_pred)

print(mae)
```

### Predictor Visualization

#### Plot Regression Line ± Error

```python
plt.figure(figsize=(12,12))

plt.rcParams.update({'font.size': 16})
plt.scatter(x, y,  color='black')
plt.plot(x, y_pred, color='royalblue', linewidth=3)
plt.plot(x, y_pred + mae, color='darkorange')
plt.plot(x, y_pred - mae, color='darkorange')
plt.title('')
plt.xlabel('Height(Inches)')
plt.ylabel('Weight(Pounds)')
plt.grid(True)
plt.show()
```

#### Calculating Pearsons Correlation Coefficient

```python
corr, pval = pearsonr(x[:,0], y[:,0])
print(corr)
```

#### Two-Tailed `p-value`

```python
print(pval < 0.05)
```
